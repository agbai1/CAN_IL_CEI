```python
import pandas as pd
import json
import folium
from geopy.geocoders import Nominatim
import requests

df = pd.read_csv('C:/Users/user/Downloads/DSI/CEI/Name_Add.csv')
json_data = df.to_json(orient='records')

# Save the JSON data to a file (optional)
with open('Name_Add.json', 'w') as f:
    f.write(json_data)
```

```python
with open('Name_Add.json') as f:
    data = json.load(f)
```

```python
addresses=[]
for i in range(100):
    addresses.append(data[i]['Address'])
```

```python
geolocator = Nominatim(user_agent="my_map")
locations = [geolocator.geocode(address) for address in addresses]
latitudes = [location.latitude for location in locations]
longitudes = [location.longitude for location in locations]

print(latitudes)
print(longitudes)
```

```python
map_center = [latitudes[0], longitudes[0]]
my_map = folium.Map(location=map_center, zoom_start=10)

# for lat, lon, address in zip(latitudes, longitudes, addresses):
#     folium.Marker([lat, lon], popup=address).add_to(my_map)

# my_map
```

```python
# Define the URL for Nominatim API
urls = [f"https://nominatim.openstreetmap.org/search?q={address}&format=json&limit=1" for address in addresses]

info=[]
for url in urls:
    # Make the request with a timeout of 5 seconds
    try:
        response = requests.get(url, timeout=5)
        response.raise_for_status()
        data = response.json()
        info.append(data)
    except requests.exceptions.RequestException as e:
        continue
    
print ("total info:", len(info))
```

```python
# Extract the latitude and longitude from the data
latitudes=[]
longitudes=[]
for data in info:
    if len(data) > 0:
        latitudes.append(data[0]["lat"])
        longitudes.append(data[0]["lon"])
        
```

```python
print(len(latitudes), len(longitudes))
```

```python
map_center = [latitudes[0], longitudes[0]]
my_map = folium.Map(location=map_center, zoom_start=10)

for lat, lon, address in zip(latitudes, longitudes, addresses):
     folium.Marker([lat, lon], popup=address, icon=folium.Icon(color="green")).add_to(my_map)

my_map
```

```python
import pandas as pd
import json
import folium
from geopy.geocoders import Nominatim
import requests
```

# Mapping MBE

```python
#OPEN COMBINED DATA DICT
with open('combined_data_MBE.json', 'r') as file:
        combined_data_MBE_df = json.load(file)
```

```python
#MBE
keysMBE=list(combined_data_MBE_df)
len(keysMBE)
```

```python
valuesMBE = list(combined_data_MBE_df.values())
len(valuesMBE)
```

```python
latMBE=[]
lonMBE=[]
namesMBE=[]
colorsMBE=[]
for i in range(len(valuesMBE)):
    if len(valuesMBE[i]) > 0:
        latMBE.append(valuesMBE[i][0]['lat'])
        lonMBE.append(valuesMBE[i][0]['lon'])
        namesMBE.append(keysMBE[i])
        colorsMBE.append('green')
    else:
        continue
print(len(latMBE), len(lonMBE), len(namesMBE), len(colorsMBE))
```

```python
latitudes=latMBE
longitudes=lonMBE
names=namesMBE
colors=colorsMBE
```

# Mapping WBE

```python
#OPEN COMBINED DATA DICT
with open('combined_data_WBE.json', 'r') as file:
        combined_data_WBE_df = json.load(file)
```

```python
len(combined_data_WBE_df)
```

```python
#WBE
keysWBE=list(combined_data_WBE_df)
len(keysWBE)
```

```python
valuesWBE = list(combined_data_WBE_df.values())
len(valuesWBE)
```

```python
latWBE=[]
lonWBE=[]
namesWBE=[]
colorsWBE=[]
for i in range(len(valuesWBE)):
    if len(valuesWBE[i]) > 0:
        latWBE.append(valuesWBE[i][0]['lat'])
        lonWBE.append(valuesWBE[i][0]['lon'])
        namesWBE.append(keysWBE[i])
        colorsWBE.append('blue')
    else:
        continue
print(len(latWBE), len(lonWBE), len(namesWBE), len(colorsWBE))
```

```python
latitudes=latitudes+latWBE
longitudes=longitudes+lonWBE
names=names+namesWBE
colors=colors+colorsWBE
```

# MAPPING

```python
map_center = [latitudes[0], longitudes[0]]
my_map= folium.Map(location=map_center, zoom_start=10)

for lat, lon, names, colors in zip(latitudes, longitudes, names, colors):
#      folium.Marker([lat, lon], popup=names, icon=folium.Icon(color=colors), icon_size=(20, 20)).add_to(my_map)
    folium.CircleMarker([lat, lon], radius=5, popup=names, color=colors, fill=True, fill_color=colors).add_to(my_map)
my_map
```

```python
my_map.save('locv2.html')
```
